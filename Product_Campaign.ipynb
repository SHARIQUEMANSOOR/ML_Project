```python
train_score=rf.predict_proba(x_train)[:,1]
real = y_train
```

```python
cutoffs = np.linspace(.001,0.999, 999)
```

```python
F2=[]
```

```python
for cutoff in cutoffs:
    predicted=(train_score>cutoff).astype(int)
    TP=((real==1)&(predicted==1)).sum()
    FP=((real==0)&(predicted==1)).sum()
    TN=((real==0)&(predicted==0)).sum()
    FN=((real==1)&(predicted==0)).sum()
    
    
    Recall=TP/(TP+FN)
    Precision=TP/(TP+FP)
    f2=(5*Precision*Recall)/((4*Precision)+Recall)
    F2.append(f2)
```

```python
import seaborn as sns
```

```python
temp=pd.DataFrame({'cutoffs':cutoffs,'F2':F2})
sns.lmplot(x='cutoffs',y='F2',data=temp,fit_reg=False)
```
